Распределение обработанных значений дохода респондентов

```python
data_income = pd.DataFrame()

data_income["Full Income R"] = data["Full Income R"]
data_income["Full Income N"] = data["Full Income N"]

data_income
```

```python
# заполнение столбца дохода респондентов, обрабатывая пропущенные значения
income_clear = []
for idx in range(data_income.shape[0]):
    if type(data_income["Full Income R"][idx]) == float:
        income_clear.append(data_income["Full Income N"][idx])
    elif type(data_income["Full Income N"][idx]) == float:
        income_clear.append(data_income["Full Income R"][idx])
    else:
        income_clear.append(np.nan)

data_income["Clear Income"] = income_clear
data_income
```

```python
data_income = data_income.drop(["Full Income N", "Full Income R"], axis = 1)
data_income = data_income["Clear Income"].dropna()

data_income
```

```python
# перевод в сформированный вид столбца clear income
clear_income_prepared = []
for income in data_income:
    if type(income) == int:
        clear_income_prepared.append(income)
    else:
        income_str = ""
        for symb in income:
            if symb.isdigit() or symb == " ":
                income_str += symb
        try:
            income_int = int(income_str.split()[0])
            if len(str(income_int)) == 3:
                income_int *= 1000
            elif len(str(income_int)) == 2:
                income_int *= 1000
            elif len(str(income_int)) == 1:
                income_int *= 1_000_000
            clear_income_prepared.append(income_int)
        except:
            clear_income_prepared.append(np.nan)

data_income = pd.DataFrame()
data_income["Clear Income Prepared"] = clear_income_prepared
data_income = data_income.dropna()
data_income
```

```python
# ГРАФИК РАСПРЕДЕЛЕНИЯ 

fig = px.histogram(data_income, color_discrete_sequence = ['#FFD600'])

fig.update_yaxes(title = "Частота", title_font = dict(size = 18, family = 'Arial', color = 'black'))
fig.update_xaxes(title = "Значение", title_font = dict(size = 18, family = 'Arial', color = 'black'))

fig.update_layout(title = "Распределение «чистых» заработков респондентов за последние 12 месяцев<br><sup><b>Учитываются все возможные источники заработка</b>",
                  font_family = "Arial",
                  font_color = "black",
                  title_font_size = 24)

# удаление подписей с boxplot`а
fig.update_yaxes(title_text='',row = 2, col = 1)
fig.update_xaxes(title_text='',row = 2, col = 1)

fig.update_layout(showlegend = False)
fig.update_layout(height = 500, width = 1000)
fig.update_layout(font_family = "Arial", font_size = 12, title_font_family = "Arial", font_color = "black")

fig.show()
```
